```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import pathlib
import json
import pandas as pd
import csv
from pathlib import Path
from scipy.stats import linregress

City_Pops = pd.read_csv("CityPopulations.csv")


City_Pops.head()
```

```python
CityPops23 = {
    'City': ['Houston, TX','Las Vegas, NV','Los Angeles, CA','Miami, FL','Phoenix, AZ'],
    'Total Population':[2589939,686443,4253574,508808,1764466]}

colors = ['red','darkorange','gold','lawngreen','blue']
    
x_axis = np.arange(len(CityPops23['City']))

plt.title('City Populations', fontsize=14)
plt.bar(CityPops23['City'], CityPops23['Total Population'], color=colors, alpha=0.5,align="center")
plt.xticks(x_axis, CityPops23['City'], rotation=45)
plt.gca().get_yaxis().set_major_formatter('{x:,.0f}')
plt.xlabel('City Choices', fontsize=12)
plt.ylabel('Total Population', fontsize=12)

for i, price in enumerate(CityPops23['Total Population']):
    plt.text(i, price, '{:,.0f}'.format(price), ha='center', va='bottom')

plt.show()
```
